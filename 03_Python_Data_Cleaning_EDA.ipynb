```python
import pandas as pd

df = pd.read_csv('customer_shopping_behaviour.csv')
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.columns
```

```python
#create a column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels) # cut the ages into four equal-sized groups based on the data distribution
```

```python
df[['age','age_group']].head(10)
```

```python
# create column purchase_frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
from sqlalchemy import create_engine
import urllib

# Step 1: Connect to MS SQL Server
# Replace 'localhost' with your actual Server Name (e.g., 'localhost\SQLEXPRESS' or your PC name)
server = "LAPTOP-098CIV5J\SQLEXPRESS" 
database = "customer_shipping_behavior"

# Create the connection string using Windows Authentication (Trusted_Connection)
# If you DO have a specific SQL username/password, remove 'Trusted_Connection=yes;' and add 'UID=your_user;PWD=your_pass;'
conn_str = (
    r'DRIVER={ODBC Driver 17 for SQL Server};'
    r'SERVER=' + server + ';'
    r'DATABASE=' + database + ';'
    r'Trusted_Connection=yes;'
    r'TrustServerCertificate=yes;'
)

# Parse string for SQLAlchemy and create the engine
quoted_conn_str = urllib.parse.quote_plus(conn_str)
engine = create_engine(f"mssql+pyodbc:///?odbc_connect={quoted_conn_str}")

# Step 2: Load DataFrame into MS SQL Server
table_name = "customer"

try:
    # This pushes your DataFrame 'df' to the SQL Server database
    df.to_sql(table_name, engine, if_exists="replace", index=False)
    print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")

except Exception as e:

    print("Error loading data:", e)
    
```
